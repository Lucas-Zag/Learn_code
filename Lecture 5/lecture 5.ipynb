```python
def f4(n):
    powers = []  # 存储 2 的幂次方
    num = n
    power = 0  # 指数

    while num > 0:
        if num & 1:  # 检查最低位是否为 1
            powers.append(f"2^{power}")  # 记录指数
        num >>= 1  # 右移，去掉最低位
        power += 1  # 指数 +1

    # 组合输出字符串
    result = f"{n} = " + " + ".join(reversed(powers))
    print(result)

# 运行示例
f4(1234567890)
```
