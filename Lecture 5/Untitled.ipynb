```python
from math import sqrt

def ok(seen_before, candidate):
    candidate_as_str = str(candidate)
    new_digits_seen = seen_before | set(candidate_as_str)
    if len(new_digits_seen) != len(seen_before) + len(candidate_as_str):
        return
    return new_digits_seen



for i in range(1, int(sqrt(997))):
    a = i * i
    _0_and_digits_in_a = ok({'0'}, a)
    if not _0_and_digits_in_a:
        continue
    for j in range(i + 1, int(sqrt(9998))):
        b = j * j
        _0_and_digits_in_a_b = ok(_0_and_digits_in_a, b)
        if not _0_and_digits_in_a_b:
            continue
        for k in range(j + 1, int(sqrt(9876532))):
            c = k * k
            _0_and_digits_in_a_b_c = ok(_0_and_digits_in_a_b, c)
            if _0_and_digits_in_a_b_c and len(_0_and_digits_in_a_b_c) == 10:
                print(a, b, c)
```

```python
import numpy as np

grid =(
       '  *     ',  
       ' ***    ', 
       '*****   ',  
       '****  * ',  
       ' **  ***',   
       '  *   * '
)



def display(*grid):
    for row in grid:
        print(' '.join(e for e in row))
   
def display_leftmost_topmost_boundary(*grid):
    grid_nb = np.array([[1 if j == '*' else 0 for j in row] for row in grid])
    size_row , size_column =grid_nb.shape
    #grids_full = np.zeros((size_row+2) * (size_column +2), int).reshape(size_row+2, size_column+2)
    grids_full = np.zeros((size_row+2 ,size_column +2), int)
    grids_full[1 : -1, 1 : -1] = grid_nb
    grids_N  = grids_full [ : -2,1 : -1]
    grids_S = grids_full[2 : , 1 : -1]
    grids_W = grids_full[1 : -1, : -2]
    grids_E = grids_full[1 : -1, 2: ]
    neighbours = grids_E + grids_W + grids_S + grids_N
    mask = (neighbours == 4 )& (grids_full[1 : -1, 1 : -1] == 1)
    grids_full[1 : -1, 1 : -1] [mask] = 0
    first_star = 0
    grid_tl = grids_full[1 : -1, 1 : -1]
    visited_nb = np.zeros((size_row ,size_column), dtype = bool)
    
    for r in range(size_row):
        for c in range(size_column):
            if grid_tl[r][c] == 1 and first_star == 0:
                first_star += 1
                stack = [(r,c)]
                first_star += 1
                while stack:
                    x, y = stack.pop()
                    if not (0 <= x < size_row and 0 <= y < size_column):
                        continue
                    if grid_tl[x][y] == 1 and not visited_nb[x][y]:
                        visited_nb[x][y] = True
                        for dr in [-1, 0, 1]:
                            for dc in [-1, 0, 1]:
                                if dr != 0 or dc != 0:
                                    stack.append((x + dr, y +dc))
    new_grid = np.array([['*' if i == 1 else " " for i in row] for row in visited_nb])

    display(*new_grid)


display_leftmost_topmost_boundary(*grid)
```
