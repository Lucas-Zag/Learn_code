```python
from solitaire1 import simulate
from collections import Counter

def main():
    n = 10000  # Number of games
    i = 42     # Starting seed

    # Run simulation
    results = simulate(n, i)  # Should return a list of remaining card counts

    # Count frequency of each result
    counter = Counter(results)

    # Compute probabilities
    output = []
    for cards_left, count in counter.items():
        prob = (count / n) * 100
        if prob >= 0.005:  # Include only strictly positive and >= 0.005%
            output.append((cards_left, prob))

    # Sort by number of cards left
    output.sort()

    # Print formatted output
    for cards_left, prob in output:
        left = f"{cards_left} cards left"
        right = f"{prob:5.2f}%"
        total_width = 45
        line = f" {left:<20} | {right:>20} "
        print(line)

if __name__ == "__main__":
    main()
```
