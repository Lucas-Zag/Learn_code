```python
from math import gcd
from math import inf
def slopes(x , y):
    #a = []
    lines = ""
    unique_set = []
    dictionary_in={}
    input_x = list(str(x))
    input_y = list(str(y))
    set_xy = list(zip(input_x,input_y))  
    [unique_set.append(item) for item in set_xy if item not in unique_set]
    #print(unique_set)
    for a,b in unique_set:
        k,l=int(b), int(a)
        if k in dictionary_in:            
            dictionary_in[k].append(l)
        else:
            dictionary_in[k] = [l]
    #print(dictionary_in)    
    new_int = [(int(a), int(b)) for a,b in unique_set ]
    new_set = sorted(new_int, key = lambda x: (-x[1] , x[0]))
    print('Here is the list of points:')
    print(new_set)
    print('')
    print('Here they are on the plane:')
    max_x = max(new_set, key = lambda x: x[0])[0]
    max_y = max(new_set, key = lambda x: x[1])[1]
    min_x = min(new_set, key = lambda x: x[0])[0]
    min_y = min(new_set, key = lambda x: x[1])[1]
    #print(max_x,max_y,min_x,min_y)
    for line in range(max_y,min_y-1,-1):
        min_nb = min_x
        lines = ''
        m = 0 
        n = 0
        if line in dictionary_in and line != min_y:
            for i in sorted(dictionary_in[line]):
                if m == 0:
                    z = i - min_nb
                    lines = str(line) + ' ' + '  '*z +'*'
                    m += 1
                    min_nb = i
                else:
                    z = i - min_nb
                    lines = lines + ' '*(1+(z-1)*2) +'*'
                    min_nb = i
            print(lines)
        elif line in dictionary_in and line == min_y:
            for i in sorted(dictionary_in[line]):
                if m == 0 and n == 0:
                    z = i - min_nb
                    lines = str(line) + ' ' + '  '*z +'o'
                    m += 1
                    n += 1
                    min_nb = i
                    global min_point_x
                    min_point_x = i
                    global min_point_y
                    min_point_y = line
                else:
                    z = i - min_nb
                    lines = lines + ' '*(1+(z-1)*2) +'*'
                    min_nb = i
            print(lines)
        else:
            print(line)
        
    x_num =' '+ ''.join(' '+ str(i) for i in range(min_x,max_x+1))
    print(x_num)
    print('')
    print('Ordered from smallest to largest, the slopes are:')
    #print(min_point_x,min_point_y)
    out_put = []
    s = []
    for (d,f) in new_set :
        if (d,f) != (min_point_x,min_point_y):
            numerator = abs(f - min_point_y)
            denumerator = abs(d - min_point_x)
            g = gcd(numerator,denumerator)
            if g > 1 and g != max(numerator,denumerator):
                numerator = numerator // g
                denumerator = denumerator// g
            if d < min_point_x:
                out_put.append(f"-{numerator}/{denumerator}")
            elif d > min_point_x and f != min_point_y:
                out_put.append(f"{numerator}/{denumerator}")
            elif d > min_point_x and f == min_point_y:
                out_put.append('0/1')
            elif d == min_point_x:
                out_put.append('inf/1')
            #print(numerator,denumerator)
            for item in out_put:
                if item not in s:
                    s.append(item)
            s.sort(key = lambda x: eval(x.replace('inf','float("inf")')))
            #s.extend( i for i in out_put if i not in s )
    print(*s, sep = ', ')
            
    #print(s)

    #print(out_put)

slopes(14627746, 20999760)
```
